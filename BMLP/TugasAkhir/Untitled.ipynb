```python
import pandas as pd

ap = pd.read_csv('_ap_v25_10%.csv', index_col = 0 )
na = pd.read_csv('_na_v25_10%.csv', index_col = 0)
eu = pd.read_csv('_eu_v25_10%.csv', index_col = 0)
br = pd.read_csv('_br_v25_10%.csv', index_col = 0)
latam = pd.read_csv('_latam_v25_10%.csv', index_col = 0)
kr = pd.read_csv('_kr_v25_10%.csv', index_col = 0)
ap_low = pd.read_csv('_ap_low_v25_100%.csv', index_col = 0)
df = pd.concat([ap,ap_low,na,eu,br,latam,kr])
df = df.reset_index()

kolom = (['nickname','region', 'rank','rank_rr','rank_rating','level', 'match', 'playtime_hours', 
                'damage_round','kill_death_ratio','headshot_rate','winrate', 'win', 'kast','damage_roun','kills','death','assist','acs',
                'kad_ratio','kill_round_ratio','first_blood','flawless_round','aces' ,'round_win',
                'agent1','agent1_hours' ,'matches1', 'win_rate1', 'k/d1', 'adr1', 'acs1', 'average_ddealt_round1', 'best_map1','map1_wr',
                'agent2', 'agent2_hours','matches2', 'win_rate2', 'k/d2', 'adr2', 'acs2', 'average_ddealt_round2', 'best_map2','map2_wr',
                'agent3', 'agent3_hours','matches3', 'win_rate3', 'k/d3', 'adr3', 'acs3', 'average_ddealt_round3', 'best_map3','map3_wr', 
                'role1', 'win_rate1', 'win_lose1', 'kda_rate1', 'kda1', 
                'role2', 'win_rate2', 'win_lose2', 'kda_rate2', 'kda2','role3', 'win_rate3', 'win_lose3', 'kda_rate3', 'kda3',
                'role4', 'win_rate4', 'win_lose4', 'kda_rate4', 'kda4','weapon_name1', 'weapon_type1', 'head_rate1', 'body_rate1', 'legs_rate1', 'kills1',
                'weapon_name2', 'weapon_type2', 'head_rate2', 'body_rate2', 'legs_rate2', 'kills2',   'weapon_name3', 'weapon_type3', 'head_rate3','body_rate3',
                'legs_rate3', 'kills3','map_name1', 'win_rate1', 'win_lose1','map_name2', 'win_rate2', 'win_lose2','map_name3', 'win_rate3', 'win_lose3',
                'map_name4', 'win_rate4', 'win_lose4', 'map_name5', 'win_rate5', 'win_lose5','map_name6', 'win_rate6', 'win_lose6','map_name7', 'win_rate7', 
                'win_lose7'])
df = df.drop('index', axis=1)
df.columns = kolom
df

    
```

```python
agent_role = pd.read_excel('agent_role.xlsx')
agent_role = agent_role.drop('no',axis=1)
agent_role['agent'] = [ i.replace('\xa0','') for i in agent_role['agent']]
agent_role['role']  = [  i.replace('\xa0','') for i in agent_role['role']]

df = df.merge(agent_role, left_on='agent1', right_on='agent', how='left').drop(columns=['agent']).rename(columns={'role':'agent1_role'})
df = df.merge(agent_role, left_on='agent2', right_on='agent', how='left').drop(columns=['agent']).rename(columns={'role':'agent2_role'})
df = df.merge(agent_role, left_on='agent3', right_on='agent', how='left').drop(columns=['agent']).rename(columns={'role':'agent3_role'})
```
